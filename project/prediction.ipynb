```python
path = os.path.join('Data','diamonds.csv')
diamonds_df = pd.read_csv(path)
```

```python
diamonds_df.drop('Unnamed: 0',axis = 1,inplace = True)
```

```python
diamonds_df['cut'].value_counts()
```

```python
X = diamonds_df.iloc[:, :-1].values
Y = diamonds_df['price']
```

```python
from sklearn.preprocessing import OneHotEncoder

from sklearn.compose import ColumnTransformer
```

```python
X
```

```python
Y
```

```python
columntransformer = ColumnTransformer([('encoder',OneHotEncoder(),[1,2,3])],remainder='passthrough')
```

```python
X = np.array(columntransformer.fit_transform(X), dtype = np.float)
X = X[:, 1:]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 300, random_state = 0)
regressor.fit(X_train, Y_train)
y_hat = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(Y_test, y_hat)) 
```
